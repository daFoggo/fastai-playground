Q4. Based on the book of the same name, what are the requirements for Parallel Distributed Processing (PDP)?
- What is PDP?
    - Parallel: Multiple neurons process information at the same time.
    - Distributed: Knowledge is stored across the network through connection weights, not in a single neuron.
    - Processing: All neurons work together to solve a task.
- Requirements of PDP
    1. A set of processing units
    - A collection of simple processing units (artificial neurons) organized into input, hidden, and output layers.
    2. A state of activation
    - Each neuron has an activation value that represents its current state or output after receiving inputs.
    3. An output function for each unit
    - Each neuron uses an activation function (e.g., ReLU, Sigmoid) to produce its output.
    4. A pattern of connectivity among units
    - Neurons are connected by weighted links that determine how information flows through the network.
    5. A propagation rule
    - A rule that describes how signals are passed from one neuron to another (forward propagation).
    6. An activation rule
    - A rule for combining the weighted inputs with the neuron's current state to compute its new output.
    7. A learning rule
    - A rule for updating the connection weights based on experience or training (e.g., backpropagation).
    8. An environment
    - The data or external environment that provides inputs for the network to learn from and operate in.

Quick Summary
1. Processing units → The neurons.
2. Activation state → Each neuron has an output value.
3. Output function → Converts the neuron's input into an output.
4. Connectivity → Weights connecting neurons.
5. Propagation rule → How signals move through the network.
6. Activation rule → How a neuron computes its output.
7. Learning rule → How the network updates weights to learn.
8. Environment → The data used for learning and prediction.

Q17. How could a feedback loop impact the rollout of a predictive policing model?
- A predictive policing model is created based on where arrests have been made in the past. In practice, this is not actually predicting crime, but rather predicting arrests, and is therefore partially simply reflecting biases in existing policing processes.
- Law enforcement officers then might use that model to decide where to focus their police activity, resulting in increased arrests in those areas.
- Data on these additional arrests would then be fed back in to retrain future versions of the model.
- This is a positive feedback loop, where the more the model is used, the more biased the data becomes, making the model even more biased, and so forth.

Q20. What is a validation set? What is a test set? Why do we need them?
- A validation set is a portion of the data that is kept separate from the training set. It is used to evaluate the model during training and to tune hyperparameters. Since the model has never seen this data during training, it provides a better estimate of how well the model generalizes to new data.
- A test set is another separate dataset that is used only after the model has been fully trained and finalized. It provides an unbiased evaluation of the model's final performance.
- We need both sets to ensure the model generalizes well to unseen data and to avoid overfitting. The validation set helps improve the model during development, while the test set gives a fair final evaluation.

Q22. Can we always use a random sample for a validation set? Why or why not?
- No. A random validation set is not always appropriate. For example, in time-series problems, the validation set should contain data that comes after the training data. Otherwise, the model may learn from future information, leading to an unrealistic evaluation.

Q24. What is a metric? How does it differ from "loss"?
- A metric is a measure used to evaluate how well a model performs on the validation set. It is designed to be easy for humans to understand, such as accuracy or error rate. (A table of metrics can be seen after a training process)
- A loss is a function used by the training algorithm to update the model's weights through gradient descent. It is optimized during training and is not necessarily easy for humans to interpret.
- In short, metrics are for evaluating a model's performance, while loss is for training the model.

Q32. What are "hyperparameters"?
- Hyperparameters are settings chosen before training that control how a model is trained. Unlike model parameters (weights), they are not learned from the data. Examples include the learning rate, batch size, number of epochs, and model architecture. Such as:
epochs = 10
learning_rate = 0.001
batch_size = 32
architecture = ResNet34
